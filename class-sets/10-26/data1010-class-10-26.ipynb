**Exercise.**  
What's the QDA decision threshold for a dataset with the given class-A and class-B observations? Use the `find_zero` function.

```julia
using Distributions, Roots
A = [2.4, 1.5, 0.2, 3.4]
B = [3.2, 4.8, 4.1]
f(x) = pdf(Normal(mean(A), std(A)), x)
g(x) = pdf(Normal(mean(B), std(B)), x)
find_zero( x -> 4/7 * f(x) - 3/7 * g(x), 4)
```

```julia
# example showing how to find square root of 2 with find_zero:
find_zero(x -> x^2 - 2, 1)
```
